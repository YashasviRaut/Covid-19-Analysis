##Task 2 : Importing the dataset and Visualizing it (Dataset 1 : Worldometer dataset)

```
from google.colab import files
files.upload()
```

```
df = pd.read_csv("covid_worldometer_data.csv")
```

```
df
```

```
# Remove commas from 'TotalCases' column
df['TotalCases'] = df['TotalCases'].str.replace(',', '')

# Convert 'TotalCases' column to numeric
df['TotalCases'] = pd.to_numeric(df['TotalCases'])

# Create the bar plot
fig = px.bar(df, x="Country", y='TotalCases', color='Country')

# Update y-axis range to start from 0
fig.update_yaxes(range=[0, df['TotalCases'].max()])

# Show the plot
fig.show()
```

```
# Convert 'TotalCases' column to numeric
df['TotalCases'] = pd.to_numeric(df['TotalCases'])

# Create the bar plot
fig = px.bar(df.head(15), x="Country", y='TotalCases', color='Country')

# Update y-axis range to start from 0
fig.update_yaxes(range=[0, df['TotalCases'].max()])

# Show the plot
fig.show()
```

```
df['TotalDeaths'] = df['TotalDeaths'].replace(' ', '0')

# Remove commas from 'TotalDeaths' column and convert to numeric
df['TotalDeaths'] = df['TotalDeaths'].str.replace(',', '').astype(int)

# Create the bar plot
fig = px.bar(df.head(15), x="Country", y='TotalDeaths', color='Country')

# Update y-axis range to start from 0
fig.update_yaxes(range=[0, df['TotalDeaths'].max()])

# Show the plot
fig.show()
```

```
df.columns
```

```
# Convert columns to numeric, coercing errors to NaN
df['Totalcases/1M'] = pd.to_numeric(df['Totalcases/1M'], errors='coerce')
df['TotalDeath/1M'] = pd.to_numeric(df['TotalDeath/1M'], errors='coerce')

# Drop rows with NaN values in either column
df.dropna(subset=['Totalcases/1M', 'TotalDeath/1M'], inplace=True)

# Create scatter plot
fig = px.scatter(df, x='Totalcases/1M', y='TotalDeath/1M', color='Country')

# Show plot
fig.show()
```

##Task 3: Geographic Visualization using Choropleth maps

```
# A choropleth map displays divided geographical areas or regions that are coloured, shaded or patterned in relation to a data variable.
#Amazing Representation of data in a map . Choropleth maps provide an easy way to visualize how a measurement varies across a geographic area.

#Dataset 1
#parameters= dataset, locations= ISOALPHA, color, hover_name, color_continuous_scale= [RdYlGn, Blues, Viridis...],
#
```

```
df.columns
```

```
px.choropleth(df,
                    locations='iso_alpha',
                    hover_name='Country',
                    color='TotalDeath/1M',
                    projection='orthographic',
                    color_continuous_scale='Viridis',
                    range_color=(0, 1000))
```

```
px.choropleth(df,
              locations = 'iso_alpha',
              hover_name = 'Country',
              color = 'TotalDeath/1M',
              projection = 'equirectangular',
              color_continuous_scale = 'Plasma',
              range_color=(0, 1000))
```

```
px.choropleth(df,
              locations = 'iso_alpha',
              hover_name = 'Country',
              color = 'TotalDeath/1M',
              scope = 'asia',
              color_continuous_scale = 'Viridis',
              range_color=(0, 1000))
```
